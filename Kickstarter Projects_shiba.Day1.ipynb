```python
#目的変数を編集する
#欠損値がある行は削除する
#df_data_dna = df_data.dropna(how='any')
```

```python
#目的変数には６つの変数が入力されている。
#今回は『成功"successful"』と『失敗"failed"』で想定してみる

#『成功"successful"』と『失敗"failed"』”だけ”抽出する
#df_data_success = df_data_dna[df_data_dna['state']=='successful']
#df_data_failed = df_data_dna[df_data_dna['state']=='failed']
df_data_success = df_data[df_data['state']=='successful']
df_data_failed = df_data[df_data['state']=='failed']

#目的変数SorFをくっつける
df_data_SorF = pd.concat([df_data_success, df_data_failed])
display(df_data_SorF)
```

```python
#目的変数を数字に置き換える
#"successful"と"failed"の2択なので0or1で変換すればいい

#ライブラリの略称
le = LabelEncoder()

#stateに出てくるカテゴリを覚える
le = le.fit(df_data_SorF['state'])

#stateを数値に変換する
df_data_SorF['state'] = le.transform(df_data_SorF['state'])
display(df_data_SorF['state'])
```
